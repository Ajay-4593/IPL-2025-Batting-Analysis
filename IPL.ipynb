```python
import pandas as pd
import matplotlib.pyplot as plt
import os


batters = pd.read_csv("IPL2025Batters.csv")

batters["HS"] = batters['HS'].str.replace('*', '', regex=False).astype(int)
#batters['HS'] = batters["HS"].astype(int)

teams = batters['Team'].nunique()

total_batters = batters['Player Name'].size

highest_sixes_by_player = batters.loc[batters['6s'].idxmax(), ['Player Name', '6s']]
#print(highest_sixes_by_player)

#--------------------top 5 sixers---------------------
top_5_sixers = batters.sort_values(['6s'],ascending = False).head(5)[['Player Name','6s']].copy()
top_5_sixers.reset_index(drop = True,inplace = True)

#--------------------top 5 fours----------------------
top_5_fours = batters.sort_values(by = '4s',ascending = False).head(5)[['Player Name','4s']]
top_5_fours.reset_index(drop =True,inplace=True)

#--------------------top 5 players with most fifties---------------------
top_5_fifties = batters.sort_values('50s',ascending = False).head(5)[['Player Name','50s']]
top_5_fifties.reset_index(drop = True,inplace= True)


least_sixes_by_player = batters.loc[batters['6s'].idxmin(), ['Player Name', '6s']]
#print(least_sixes_by_player)

total_sixes_per_team = batters.groupby('Team')['6s'].sum()

total_sixes_per_team = total_sixes_per_team.to_frame()
total_sixes_per_team = total_sixes_per_team.sort_values(['6s'],ascending = False)
#print(highest_sixes_per_team)

indexes = total_sixes_per_team.index
#print(indexes)

#-----------------------------least sixes scored by each team---------------------------------
least_sixes_team_name = total_sixes_per_team.idxmin()
least_sixes_team_score = total_sixes_per_team.min()
#print(least_sixes_team_name, end='-')
#print(least_sixes_team_score)

#-----------------------------Highest sixes scored by each team---------------------------------
highest_sixes_team_name = total_sixes_per_team.idxmax()
highest_sixes_team_score = total_sixes_per_team.max()
#print(highest_sixes_team_name, end='-')
#print(highest_sixes_team_score)

#-----------------------------------highest score in a single Innings---------------------------------
highest_score_of_season = batters.sort_values(['HS'],ascending = False).head(5)[['Player Name','HS']].copy()
highest_score_of_season.reset_index(drop =True,inplace = True)
#print(highest_score_of_season)
#print(highest_score_of_season['Player Name'])

min_balls = 200
min_balls_played = batters[batters['BF'] > min_balls]
#print(min_balls_played)

highest_strike_by_player = min_balls_played.loc[min_balls_played['SR'].idxmax(), ['Player Name', 'SR']]

top_strike_rate = min_balls_played.sort_values(["SR"],ascending = False).head(5)[['Player Name','SR']].copy()
top_strike_rate.reset_index(drop = True,inplace =True)
#print(highest_strike_by_player)

#most_runs_of_season =
# print(f'leading six hitter of the season is {highest_sixes_by_player['Player Name']} with {highest_sixes_by_player['6s']} 6`s`')
lines = '-----------------------------------------------------------------------------------------------------------------------------'
print(f"Top 5 players with most sixers:-\n{top_5_sixers.loc[0,'Player Name']} holds the record for the highest six hitter of the IPL season 2025 with {top_5_sixers.loc[0,'6s']} 6`s\n Here are the Top 5 players with most sixes\n")
print(f"{top_5_sixers}\n ")

print(f"{lines}\n top 5 players with most fours:-\n{top_5_fours.loc[0,'Player Name']} holds the record of highest fours hitter of the IPL season 2025 with {top_5_fours.loc[0,'4s']} 4`s\n Here are the Top 5 players with most fours\n ")
print(f"{top_5_fours}\n")

print(f"{lines}\n top 5 players with most fifties:-\n{top_5_fifties.loc[0,'Player Name']} the `GREATEST OF ALL TIME` set the record for most number of fifties the IPL season 2025 with {top_5_fifties.loc[0,'50s']} 50`s\n Here are the Top 5 players with most fifties\n")
print(f"{top_5_fifties}\n")
print(f"{lines}\nSixes scored by each team:-\nMost number of sixes scored by {highest_sixes_team_name["6s"]} with {highest_sixes_team_score["6s"]} 6`s in the IPL season 2025 which is absolutely phenomenal\n Here are total sixes scored by each team:-")
#------------------------------Either you can use for loop or list comprehension----------------------------------
for i in indexes:
    print(f"{i} scored {total_sixes_per_team.loc[i,'6s']}")

# sixes_team = [f"{i} scored {total_sixes_per_team.loc[i,'6s']}" for i in indexes]
# print(sixes_team)

print(f"{lines}\n In 2025 {least_sixes_team_name["6s"]} struggled to deliver sixes ending as the least sixes scored in the season with only {least_sixes_team_score["6s"]} 6`s throughout the season")

# print(highest_score_of_season)
print(f"{lines}\n top 5 batsmen highscores in a single innings\n{highest_score_of_season.loc[0,'Player Name']} is also called as `PULI` secured highest score of the season with phenomenal batting skills with {highest_score_of_season.loc[0,"HS"]} in a single innings\nPresenting you highest score in a innings:-\n{highest_score_of_season}")

print(f"{lines}\nTop 5 batsmen with top strike Rate with minimum 200 balls faced\n{top_strike_rate.loc[0,"Player Name"]} has played aggressively throughout the season with a strike rate of {top_strike_rate.loc[0,'SR']}\n\n listing you top 5 players who played aggressively with top strike Rate:-\n{top_strike_rate}")

print(f"{lines} Top 5 batsmen with most runs in the season\n {batters.loc[0,"Player Name"]} performed consistently overall the season showed his absolute batting skills and scored {batters.loc[0,"Runs"]} runs`s\n\n Here are your Top Runs scorers of the IPL season 2025:-\n{batters.loc[0:4,['Player Name','Runs']]}")

print(lines+'\n')
avg_sixes = total_sixes_per_team['6s'].mean()
for team in total_sixes_per_team.index:
    sixes = total_sixes_per_team.loc[team,'6s']
    if sixes > avg_sixes :
        print(f'{team} mostly depends on six hitting with {sixes} sixes')
    else:
        print(f'{team} shows a balanced approach with descent six hitting with {sixes} sixes')





fig, axes = plt.subplots(2, 2, figsize=(12, 8))

top_5_sixers.plot(x='Player Name', y='6s', kind='bar',ax=axes[0, 0],legend = False)
axes[0, 0].set_title('Top 5 Six Hitters')
axes[0, 0].set_xlabel('Players')
axes[0, 0].set_ylabel('Sixes')
axes[0, 0].grid()

top_5_fifties.plot(x='Player Name',y = '50s',kind = 'bar',ax = axes[0,1],legend = False)
axes[0,1].set_title('Top 5 Fifties scorers')
axes[0,1].set_xlabel('Players')
axes[0,1].set_ylabel('Fifties')
axes[0,1].grid()

top_strike_rate.plot(x='Player Name',y='SR',kind = 'bar',ax = axes[1,0],legend = False)
axes[1,0].set_title('Top 5 highest strike rate players')
axes[1,0].set_xlabel('players')
axes[1,0].set_ylabel('Strike rate')
axes[1,0].grid()

total_sixes_per_team.plot(kind = 'bar',ax = axes[1,1])
axes[1,1].set_title('All teams overall sixes comparision')
axes[1,1].set_xlabel('Team Name')
axes[1,1].set_ylabel('Sixes scored by each team')
axes[1,1].grid()

axes[0,0].tick_params(axis='x', rotation=45)
axes[0,1].tick_params(axis='x', rotation=45)
axes[1,0].tick_params(axis='x', rotation=45)
axes[1,1].tick_params(axis='x', rotation=45)

for ax in axes.flat:
    for container in ax.containers:
        ax.bar_label(container)


plt.suptitle("IPL 2025 batsmen analysis dashboard",fontstyle = 'oblique',fontsize = 15)
#plt.tight_layout()

for ax in axes.flat:
    ax.grid(axis='y', linestyle='--', alpha=1)

for ax in axes.flat[:3]:
    bars = ax.containers[0]
    values = [bar.get_height() for bar in bars]
    max_index = values.index(max(values))
    bars[max_index].set_color('Green')


bars = axes[1,1].containers[0]
values = [bar.get_height() for bar in bars]
bars[values.index(max(values))].set_color('Green')

plt.tight_layout(rect=[0, 0, 1, 0.95])
```
